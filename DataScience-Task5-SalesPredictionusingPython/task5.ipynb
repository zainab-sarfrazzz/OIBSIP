```python
x = df[['Newspaper', 'Radio', 'TV']]
y = df['Sales']
```

**Model performance:** The following terms are used to test the model performance and its very imported to check how well your model perfromed:


*   Linear_regression
*   Training
*   Testing
*   Predictions



```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
LR = LinearRegression()
LR.fit(x_train,y_train)
```

```python
pred = LR.predict(x_test)
```

```python
mae = mean_absolute_error(y_test, pred)
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, pred)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

```python
importance = pd.Series(LR.coef_, index=x.columns)
importance.plot(kind='bar')
plt.title('Feature Importance')
plt.show()
```

```python
residuals = y_test - pred
plt.scatter(pred, residuals)
plt.axhline(y=0, color='y', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```

**Conclusions:**

*   TV advertising has the strongest impact on sales.
*   Radio also contributes positively.
*   Newspaper has the least influence.
*   Random Forest performed better than Linear Regression
